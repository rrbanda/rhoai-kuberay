# Submit a RayJob with the CodeFlare SDK

This notebook demonstrates how to submit batch Ray jobs using the CodeFlare SDK.

**Prerequisites:**
- Running in an RHOAI Workbench
- `ray-demo` namespace exists with Kueue configured
- KubeRay operator is running

```python
from codeflare_sdk import RayJob, ManagedClusterConfig
```

## Workflow 1: Ephemeral RayJob

This creates a temporary Ray cluster, runs the job, and tears down the cluster automatically.

```python
cluster_config = ManagedClusterConfig(
    num_workers=2,
    worker_cpu_requests=1,
    worker_cpu_limits=1,
    worker_memory_requests=2,
    worker_memory_limits=2,
)

job = RayJob(
    job_name="ephemeral-demo",
    entrypoint="python -c \"import ray; ray.init(); print('Resources:', ray.cluster_resources()); print('SUCCESS!')\"",
    cluster_config=cluster_config,
    namespace="ray-demo",
    local_queue="default",
)

job.submit()
```

```python
job.status()
```

```python
job.logs()
```

## Workflow 2: Job on Existing Cluster

Submit a job to a running RayCluster for fast iteration (no cluster startup wait).

**Prerequisite:** A running cluster named `demo-cluster` in `ray-demo` namespace.

```python
quick_job = RayJob(
    job_name="quick-test",
    entrypoint="python -c \"import ray; ray.init(); print('Nodes:', len(ray.nodes())); print('Quick test done!')\"",
    cluster_name="demo-cluster",
    namespace="ray-demo",
)

quick_job.submit()
```
